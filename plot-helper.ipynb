```python
import matplotlib.pyplot as plt
```

```python
text_file1 = open("acc1.txt", "r")
lines1 = text_file1.readlines()
lines1 = [float(i) for i in lines1]
text_file2 = open("acc2.txt", "r")
lines2 = text_file2.readlines()
lines2 = [float(i) for i in lines2]
```

```python
print(type(lines1[0]))
```

```python
x_ = []
for i in range(len(lines1)):
    x_.append(i)
plt.plot(x_, lines1, label="without augmentation")
plt.plot(x_, lines2, label="with augmentation")
plt.xlim([0.0, len(lines1) - 1])
plt.ylim([min(lines1) - 0.05, max(lines2) + 0.05])
plt.xlabel('Epochs')
plt.ylabel('val_acc')
plt.legend(loc="lower right")
plt.show()
```

```python
def readfile(filename):
    text_file = open(filename, "r")
    lines = text_file.readlines()
    lines = [float(i) for i in lines]
    return lines
```

```python
highlr_highreg = readfile("highlr_highreg.txt")
highlr_lowreg = readfile("highlr_lowreg.txt")
plt.plot(x_, highlr_highreg, label="highlr_highreg")
plt.plot(x_, highlr_lowreg, label="highlr_lowreg")
plt.xlim([0.0, len(highlr_highreg) - 1])
plt.ylim([min(highlr_highreg) - 0.05, max(highlr_highreg) + 0.05])
plt.xlabel('Epochs')
plt.ylabel('val_acc')
plt.legend(loc="lower right")
plt.show()
```

```python
lowlr_highreg = readfile("lowlr_highreg.txt")
lowlr_lowreg = readfile("lowlr_lowreg.txt")
plt.plot(x_, lowlr_highreg, label="lowlr_highreg")
plt.plot(x_, lowlr_lowreg, label="lowlr_lowreg")
plt.xlim([0.0, len(highlr_highreg) - 1])
plt.ylim([min(highlr_highreg) - 0.05, max(highlr_highreg) + 0.05])
plt.xlabel('Epochs')
plt.ylabel('val_acc')
plt.legend(loc="lower right")
plt.show()
```

```python
nodecay = readfile("nodecay.txt")
decay = readfile("decay.txt")
plt.plot(x_, nodecay, label="nodecay")
plt.plot(x_, decay, label="decay")
plt.xlim([0.0, len(nodecay) - 1])
plt.ylim([min(nodecay) - 0.05, max(decay) + 0.05])
plt.xlabel('Epochs')
plt.ylabel('val_acc')
plt.legend(loc="lower right")
plt.show()
```
